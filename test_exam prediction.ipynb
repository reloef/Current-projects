```python
#-------------------Loading data from my destop-----------------------------------

df = pd.read_csv(r'C:\Users\reloef\Documents\hippo powerbi\soial_data1.csv')
df = pd.DataFrame(df, index = None)

df.head()
```

```python
df.describe()
```

```python
#--------Checking for percentage of null values on every column------------#

print(round((df.isnull().sum()/df.shape[0])*100,2).astype(str)+ '%')
```

```python
df.info()
```

```python
df['classification'] = df['classification'].map({'very comfortable':2, 'comfortable':1,'neutral':0,'uncomfortable':-1,'very uncomfortable':-2 })
```

```python
df.head()
```

```python
col = "classification"
dlx = df.loc[:, df.columns != col]
dly = list(dlx.columns)
print(dly)
```

```python
X = df[['funding', 'housing', 'stipend p/m', 'matric carriculum', 'home language', 'english proficiancy ', 'computer use', 'social media hours', 'study time hrs p/d', 'smart phone', 'work during studies', 'APS score', 'maths symbol', 'physics symbol', 'english symbol', 'distance traveled']]
y = df['classification']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .25,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
y_pred = rfc.predict(X_test)
print(classification_report(y_test,y_pred))
```
